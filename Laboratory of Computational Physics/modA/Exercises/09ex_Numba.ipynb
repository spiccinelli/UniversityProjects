## Exercise 1

Create a vectorized version of log and exp math function for 1D array A = [2, 5, 10, 3, 8]

Results should be: 
+ [0.6931472 1.609438  2.3025851 1.0986123 2.0794415]
+ [7.3890562e+00 1.4841316e+02 2.2026465e+04 2.0085537e+01 2.9809580e+03]

````python
A = [2, 5, 10, 3, 8]

@vectorize([float32(int64)], target='parallel', fastmath=True)
def log(a):
    c =  np.log(a)
    return c
@vectorize([float32(int64)], target='parallel', fastmath=True)
def exp(a):
    c = np.exp(a)
    return c

print(log(A), exp(A))
````

Compute the value of a Gaussian probability density function at $x$ with $mean = 1$, $\sigma = 1$, lower and upper bound in $(-3, 3)$ and $size = 100000$

````python
@jit
def gauss_pdf(n = 100000):
    a = np.array([(np.random.normal(loc=1.0, scale=1.0)-3)/3 for i in range(n)])
    return np.interp(a, (a.min(), a.max()), (-3, +3)) #to scale to the [-3, 3] interval

gauss_pdf()
````

## Exercise 3

Create a "zero suppression" function. A common operation when working with waveforms is to force all samples values below a certain absolute magnitude to be zero, as a way to eliminate low amplitude noise. 
Plot the data before and after the application of the zero_suppress function.

$thresold = 15$

````python
n = 100000
noise = np.random.normal(size=n) * 3
pulses = np.maximum(np.sin(np.arange(n) / (n / 23)) - 0.3, 0.0)
data = ((pulses * 300) + noise).astype(np.int16)

@vectorize(nopython=True)
def zero_suppression(data, threshold):
    if np.abs(data) >= threshold: return data
    else: return 0.0

fig, axes = plt.subplots(nrows=1, ncols=2, tight_layout=False, figsize=(15, 6), sharey=True)
fig.subplots_adjust(wspace=0.1)
ax0, ax1 = axes.flatten()
ax0.set_xlabel(r"Data before zero_suppression func.", size=16)
ax1.set_xlabel(r"Data after zero_suppression func.", size=16)
ax0.set_ylabel(r"Signal", size=16)
ax0.plot(data, color = 'red')
ax1.plot(zero_suppression(data, 15), color = 'green')
plt.show()
````

$k(x,y) = \tanh(\alpha x^T y + c) $

## Exercise 5

Create a kernel function similar to the ```double_kernel``` see during the lecture and create a new function that takes a 3 dimensional matrix as input calculating the $cos$ for each element and then returns the result. The shape of the matrix must be $256X256X256$. The matrix can be randomly generated

````python
@cuda.jit
def double_kernel_3D(io_array):
    row, col, dep = cuda.grid(3)
    if row < io_array.shape[0] and col < io_array.shape[1] and dep < io_array.shape[2]:
        io_array[row][col][dep] = np.cos(io_array[row][col][dep])
        
data = np.random.rand(256,256,256)

threadsperblock = (16, 8, 2)
blockspergrid_x = int(math.ceil(data.shape[0] / threadsperblock[0]))
blockspergrid_y = int(math.ceil(data.shape[1] / threadsperblock[1]))
blockspergrid_z = int(math.ceil(data.shape[2] / threadsperblock[2]))

blockspergrid = (blockspergrid_x, blockspergrid_y, blockspergrid_z)
double_kernel_3D[blockspergrid, threadsperblock](data)

print(data)
````
